# Loss Landscape Slice

Neural networks have many parameters, so we cannot plot the full loss surface. Instead, this notebook plots a 2D slice around a trained MNIST feedforward model.

## Initialize PyTorch

- Import tools for copying model weights, training, and plotting
- Seed PyTorch for reproducible random directions
- Detect device (GPU, Apple Silicon, CPU)

```python
import copy
import torch
from torch import nn
from torch.utils.data import DataLoader, Subset
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

torch.manual_seed(42)

def get_device():
    if torch.cuda.is_available():
        return "cuda"
    if torch.backends.mps.is_available():
        return "mps"
    return "cpu"

device = get_device()
```

## Load Landscape Data

- Load MNIST train and test data
- Use a training subset to fit the model quickly
- Select one test batch for evaluating the loss landscape

```python
transform = transforms.Compose([transforms.ToTensor(), transforms.Normalize((0.1307,), (0.3081,))])
train_data = datasets.MNIST(root="../../data", train=True, download=True, transform=transform)
test_data = datasets.MNIST(root="../../data", train=False, download=True, transform=transform)
train_loader = DataLoader(Subset(train_data, range(6000)), batch_size=128, shuffle=True)
landscape_loader = DataLoader(Subset(test_data, range(256)), batch_size=256)
landscape_images, landscape_labels = next(iter(landscape_loader))
landscape_images, landscape_labels = landscape_images.to(device), landscape_labels.to(device)
```

## Define Model

- Create a small feedforward model factory
- Use the same architecture for training and landscape evaluation
- Set up cross-entropy loss and Adam optimizer

```python
def build_model():
    return nn.Sequential(
        nn.Flatten(),
        nn.Linear(28 * 28, 128),
        nn.ReLU(),
        nn.Linear(128, 64),
        nn.ReLU(),
        nn.Linear(64, 10),
    ).to(device)

model = build_model()
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
```

## Train Reference Model

- Train the model for several epochs
- Save a copy of the trained weights
- Use the trained weights as the center of the landscape slice

```python
for epoch in range(3):
    model.train()
    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)
        loss = loss_fn(model(images), labels)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

trained_state = copy.deepcopy(model.state_dict())
```

## Create Landscape Directions

- Sample two random directions in parameter space
- Scale each direction to match the size of each parameter tensor
- Define helper functions to move away from trained weights
- Evaluate batch loss at any point on the 2D slice

```python
def random_direction_like(state):
    direction = {}
    for name, value in state.items():
        d = torch.randn_like(value)
        d = d / (d.norm() + 1e-8) * (value.norm() + 1e-8)
        direction[name] = d
    return direction

direction_a = random_direction_like(trained_state)
direction_b = random_direction_like(trained_state)

def state_at(alpha, beta):
    return {
        name: trained_state[name] + alpha * direction_a[name] + beta * direction_b[name]
        for name in trained_state
    }

def batch_loss_at(alpha, beta):
    model.load_state_dict(state_at(alpha, beta))
    model.eval()
    with torch.no_grad():
        return loss_fn(model(landscape_images), landscape_labels).item()
```

## Sample Loss Grid

- Create a grid of coordinates around the trained weights
- Evaluate loss at each point in the two random directions
- Restore the trained weights after sampling

```python
grid = torch.linspace(-0.7, 0.7, 25)
losses = torch.zeros(len(grid), len(grid))

for i, alpha in enumerate(grid):
    for j, beta in enumerate(grid):
        losses[i, j] = batch_loss_at(alpha.item(), beta.item())

model.load_state_dict(trained_state)
```

## Plot Loss Contours

- Render the sampled loss values as a contour map
- Mark the trained weights at the center point
- Use the plot to discuss local valleys and high-loss regions

```python
plt.figure(figsize=(7, 6))
contours = plt.contourf(grid, grid, losses, levels=30, cmap="viridis")
plt.colorbar(contours, label="cross entropy loss")
plt.scatter([0], [0], color="red", label="trained weights")
plt.xlabel("direction A")
plt.ylabel("direction B")
plt.title("2D slice through the loss landscape")
plt.legend()
plt.tight_layout()
```

## Discussion

This plot is not the full loss landscape. It is one slice through a high-dimensional parameter space. Try changing the grid range, model size, or training epochs to see how the slice changes.
