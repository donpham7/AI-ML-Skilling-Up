# Feedforward Neural Network Quickstart

This notebook introduces tensors, datasets, a small feedforward neural network, and a minimal training loop.

## Setup

In Google Colab, PyTorch is usually preinstalled. If you are running locally, install dependencies from the repository root with `pip install -r requirements.txt`.

```python
import torch
from torch import nn
from torch.utils.data import DataLoader, TensorDataset

torch.manual_seed(42)
```

```python
X = torch.randn(256, 2)
y = ((X[:, 0] + X[:, 1]) > 0).long()

dataset = TensorDataset(X, y)
loader = DataLoader(dataset, batch_size=32, shuffle=True)
```

```python
model = nn.Sequential(
    nn.Linear(2, 8),
    nn.ReLU(),
    nn.Linear(8, 2),
)

loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
```

```python
for epoch in range(20):
    total_loss = 0.0

    for batch_X, batch_y in loader:
        logits = model(batch_X)
        loss = loss_fn(logits, batch_y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    if (epoch + 1) % 5 == 0:
        print(f"epoch {epoch + 1}: loss={total_loss / len(loader):.4f}")
```

```python
with torch.no_grad():
    predictions = model(X).argmax(dim=1)
    accuracy = (predictions == y).float().mean()

accuracy.item()
```
